# 📊 Rayda Fleet Telemetry Data Exploration & Ad-Hoc Analytics
This workbook runs data checks directly against the localized DuckDB time-series storage engine to verify performance trends before running the agent graph loops.

```python
import duckdb
import pandas as pd

# Connect to the compiled analytical database storage layer
conn = duckdb.connect('../data/fleet_telemetry.db')
print("Database Connection Established Securely.")
```

### 🔍 Ad-Hoc Check 1: 30-Day Device Battery Cycle Aggregations

```python
query_battery = """
SELECT device_id, company_id, max(battery_cycle_count) as peak_cycles, max(battery_condition) as current_state
FROM snapshots
GROUP BY device_id, company_id
ORDER BY peak_cycles DESC;
"""
df_battery = conn.execute(query_battery).fetchdf()
df_battery.head()
```

### 🔍 Ad-Hoc Check 2: Core Memory Saturation Tracking

```python
query_ram = """
SELECT device_id, company_id, AVG(ram_used_pct) as mean_ram_saturation
FROM snapshots
GROUP BY device_id, company_id
HAVING mean_ram_saturation > 80.0;
"""
df_ram = conn.execute(query_ram).fetchdf()
df_ram.head()
```
